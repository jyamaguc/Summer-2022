# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder as ohe
```

# Cleaning

```python
df = pd.read_csv(r'C:\Users\jyama\Git\Lung_cancer\survey lung cancer.csv')
df.head()
```

```python
def convert(ans):
    if ans =="YES":
        return 1
    else:
        return 0
```

```python
df["LUNG_CANCER"] = df['LUNG_CANCER'].apply(convert)
```

```python
df = df.dropna()
df.shape
```

```python
#one hot encode gender
ohe = ohe()
x = pd.DataFrame(ohe.fit_transform(df[['GENDER']]).toarray())
```

```python
df['F'], df['M'] = x[0], x[1]
df = df.drop(columns = ['GENDER'])
df.head()
```

```python
#split dataframe into data and target
data = df.drop(columns =['LUNG_CANCER'])
target = pd.DataFrame(df['LUNG_CANCER'])
target
```

# Model Selection

```python
from sklearn import tree
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.linear_model import LogisticRegression
```

```python
kf = KFold(n_splits = 10)
```

```python
#get model score
def model_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)

tree_score=[]
knn_score = []
rf_score = []
lr_score = []

for train_index, test_index in kf.split(data):
    X_train, X_test, y_train, y_test = data.iloc[train_index], data.iloc[test_index], target.iloc[train_index], target.iloc[test_index] 
    knn_score.append(model_score(KNN(n_neighbors = 7), X_train, X_test, y_train, y_test))
    tree_score.append(model_score(tree.DecisionTreeClassifier(), X_train, X_test, y_train, y_test))
    rf_score.append(model_score(RF(), X_train, X_test, y_train, y_test))
    lr_score.append(model_score(LogisticRegression(), X_train, X_test, y_train, y_test))
```

```python
#finding mean of all model scores
from statistics import mean
def get_mean(score):
    return(mean(score))
```

```python
scores = [lr_score, rf_score, tree_score, knn_score]
for item in scores:
    print('Score = '+str(get_mean(item)))
#random Forest seems to be the best model.
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV as GSCV
```

```python
max_features = np.arange(1,17,1)
n_estimators = np.arange(10,210,10)

param_grid = dict(max_features= max_features, n_estimators = n_estimators)

grid = GSCV(estimator = RF(), param_grid = param_grid, cv =5)
```

```python
grid.fit(X_train, y_train)
```

```python
print('The best parameters are %s with score of %f' %(grid.best_params_, grid.best_score_))
```

# Final Model

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.2)
```

```python
#Random forest model
rf = RF(max_features = 2, n_estimators = 40).fit(X_train, y_train)
score = rf.score(X_test, y_test)
```
